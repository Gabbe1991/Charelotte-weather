```python
import pandas as pd
import json
with open('accidents.json') as f:
    data = json.load(f)

#print(df.head())
```

```python
df = pd.json_normalize(data)
df.sample(10)
df.info()
```

```python
#df.drop(columns = ["_id","event_no"], axis= 1, inplace= True)
#df.info()
df["weatherInfo.coord.lat.$numberInt"].sample(10)
```

```python
missing_data = df.isnull().mean() * 100
```

```python
missing_data = missing_data[missing_data > 0].sort_values(ascending = False)
print(missing_data)

#The numbers represent that % amount missing in each column, 
#Im gonna remove anything above 5% missing values.
```

```python
missing_percentage = df.isnull().mean() * 100
df = df.loc[:, missing_percentage <= 5]
```

```python
df.info()
```

```python
df.sample(20)
```

```python
df.loc[:, 'weatherInfo.dt'] = pd.to_datetime(df['weatherInfo.dt.$numberInt'].astype(float), unit='s')
df.loc[:, 'weatherInfo.sys.sunrise'] = pd.to_datetime(df['weatherInfo.sys.sunrise.$numberInt'].astype(float), unit='s')
df.loc[:, 'weatherInfo.sys.sunset'] = pd.to_datetime(df['weatherInfo.sys.sunset.$numberInt'].astype(float), unit='s')
```

```python
df.drop(columns=[
    'weatherInfo.dt.$numberInt', 
    'weatherInfo.sys.sunrise.$numberInt', 
    'weatherInfo.sys.sunset.$numberInt', 
    'weatherInfo.main.temp_min.$numberDouble', 
    'weatherInfo.main.temp_max.$numberDouble', 
    'weatherInfo.weather', 
    'weatherInfo.base'
], errors='ignore', inplace=True)

df.rename(columns={
    'weatherInfo.id.$numberInt': 'weather_station_id',
    'weatherInfo.name': 'weather_location_name',
    'weatherInfo.cod.$numberInt': 'weather_status_code',
    'weatherInfo.sys.id.$numberInt': 'weather_system_id',
    'weatherInfo.sys.type.$numberInt': 'weather_system_type',
    'weatherInfo.visibility.$numberInt': 'weather_visibility_KM',  # after conversion to KM
    'weatherInfo.main.pressure.$numberInt': 'weather_pressure_hPa',
    'weatherInfo.dt': 'weather_datetime',
    'weatherInfo.sys.sunrise': 'sunrise_time',
    'weatherInfo.sys.sunset': 'sunset_time'
}, inplace=True)
# Convert the column to numeric type (float) and then divide by 1000
df['weather_visibility_KM'] = pd.to_numeric(df['weather_visibility_KM'], errors='coerce') / 1000

```
